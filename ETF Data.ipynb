```python
import yfinance as yf
import pandas as pd
```

```python
# Extract indices' daily close price
start = '2006-01-01'
end = '2020-12-31'
tickers = 'VTI AGG DBC ^VIX'.split()
adj_close = yf.download(' '.join(tickers), start=start, end=end)['Adj Close'].dropna()
adj_close
```

```python
# Calculate the daily return
for tic in list(tickers):
  adj_close[tic+'_Return'] = ((adj_close[tic] / adj_close[tic].shift(1)-1))

data = adj_close.dropna()
data
```

```python
data = data.rename(columns={
    "AGG": "AGG_Price", "DBC": "DBC_Price", "VTI": "VTI_Price",
    "^VIX": "VIX_Price", "^VIX_Return": "VIX_Return"})

data.head()
```

```python
ret_data = data[['VTI_Price', 'VTI_Return', 'AGG_Price', 'AGG_Return', 'DBC_Price', 'DBC_Return', 'VIX_Price', 'VIX_Return']]
ret_data.head()
```
